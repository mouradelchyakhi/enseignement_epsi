```python
import duckdb
con = duckdb.connect()

df_duckdb = con.execute("SELECT * FROM df LIMIT 10").df()
```

```python
df_duckdb_unicity = con.execute(
    """ 
    SELECT product_id, product_name, category,price, stock_quantity, date_added, COUNT(*) as count
    FROM df
    GROUP BY product_id, product_name, category,price, stock_quantity, date_added
    HAVING COUNT(*) > 1
    ORDER BY 1 DESC
    """).df()
```

```python
df_duckdb_unicity.head()
```

```python
# Pour supprimer les doublons en gardant la première occurrence, on peut utiliser la fonction ROW_NUMBER() qui est une fonction fenêtrée

con.execute("""
            SELECT *, ROW_NUMBER() OVER (partition by product_id) as row_number_product_id 
            FROM df 
            WHERE product_id IN ('P1059','P1076')
            """).df()
```

```python
df_corrected = con.execute("""
            WITH CTE AS (
            SELECT *, ROW_NUMBER() OVER (partition by product_id) as row_number_product_id
            FROM df
            )

            SELECT *
            FROM CTE
            WHERE row_number_product_id = 1 
            """).df()
```

```python
# Complétude

con.execute("""
            SELECT *
            , COALESCE(stock_quantity, 0) as stock_quantity_complete, 
            CASE WHEN COALESCE(price, '') = '' THEN 1 ELSE 0 END AS price_missing_flag       
            FROM df_corrected 
            WHERE stock_quantity IS NULL OR price IS NULL
            """).df()

df_corrected = con.execute("""
            SELECT *
            , COALESCE(stock_quantity, 0) as stock_quantity_complete, 
            CASE WHEN COALESCE(price, '') = '' THEN 1 ELSE 0 END AS price_missing_flag       
            FROM df_corrected 
            WHERE price_missing_flag = 0
            """).df()
```

```python
# nous vérifions que nous avons bien supprimé les lignes où le prix est manquant
df_corrected[df_corrected['price_missing_flag'] == 1]
```

```python
# Format et type

print("Types avant conversion : \n" , df_corrected.dtypes)

df_corrected = con.execute("""
            SELECT * , CAST( REPLACE (TRIM(REPLACE(price,'€',' ')), ',','') AS double) AS price_cleaned
            FROM df_corrected 
            """).df()

print("Types après conversion : \n" , df_corrected.dtypes)

```

```python
# Validité

con.execute("""
            SELECT stock_quantity, CASE WHEN stock_quantity < 0 THEN 0 ELSE stock_quantity END AS stock_quantity_valid
            FROM df_corrected 
            WHERE stock_quantity < 0
            """).df().head()
```

```python
# Cohérence

con.execute("""
            SELECT DISTINCT category, REPLACE (LOWER(category), 'é','e') AS category_lowercase
            FROM df_corrected 
            """).df()
```

```python
# Fraicheur

con.execute("""
            SELECT substring(date_added,1,4) as date_added_year, count(*) as count_records
            FROM df_corrected 
            group by date_added_year
            """).df()
```

```python
import pandas as pd
df = pd.read_csv('Td2Exercice2.csv')
df
```

```python
# re est une librairie permettant de manipuler les expressions régulières

import re
```

```python
# Astuce: testez vos Regex avec https://regex101.com/
# Ressource pour apprendre les Regex : https://regexlearn.com/cheatsheet


# je teste sur mon adresse email le pattern défini à l'aide de https://regex101.com/

s = "mourad.elchyakhi@gmail.com."
print(re.match(r"^\w+\.\w+\@.\w+\.(fr|com)", s))
```

```python
# nous vérifions que chaque email correpsond bien à un pattern d'email

df['flag_email_pattern'] = df['Email'].apply(lambda x : 1 if re.match(r"^\w+\.\w+\@.\w+\.(fr|com|org)", x) else 0)
df
```

```python
# je formatte les numéros afin d'avoir une valeur formattée. Attention à l'ordre des remplacements qui est très important.
df['Telephone_FR'] = df['Telephone_FR'].str.replace('.','').str.replace('Tel: ','').str.replace('+33(0)','0').str.replace('+33','0').str.replace(' ','').str.replace('-','')
df
```

```python
# Je vérifie à l'aide d'une expression régulière que le numéro de téléphone comporte bien 10 caractère, le tout via un flag 0 / 1 
df['Telephone_FR_formatted_flag'] = df['Telephone_FR'].apply(lambda x : 1 if re.match(r"\d{10}", x) else 0)
df
```

```python
df.dtypes
```

```python
re.sub(r"(\d{2})", r"\1 ",'065010').strip()
```

```python
# Enfin à l'aide d'une expression régulière je formatte le numéro de téléphone, et si le flag est 0 alors j'indique que le numéro de téléphone n'est pas renseigné

lambda_function = lambda x: re.sub(r"(\d{2})", r"\1 ", x['Telephone_FR']).strip() if x['Telephone_FR_formatted_flag'] == 1 else 'Non renseigné'

df['Telephone_FR_formatted'] = df.apply(lambda_function, axis=1)
df
```

```python
# à l'aide d'une expression regulière je retire les points et les virgules
lambda_function = lambda x: re.sub(r"[\.,]", r" ", x['Nom_Complet']).strip()
df['Nom_Complet_corr'] = df.apply(lambda_function, axis=1)
df
```

```python
# A l'aide d'une seconde expression régulière je sépare le pattern "EveMarie" c'est à dire un mot attaché avec une lettre majuscule qui n'est pas la première lettre

lambda_function2 = lambda x: re.sub(r"([a-z])([A-Z])", r"\1 \2", x['Nom_Complet_corr']).strip()
df['Nom_Complet_corr'] = df.apply(lambda_function2, axis=1)
df
```

```python
# A l'aide d'une dernière fonction, je transforme tous les noms en nom propre (uniquement première lettre en majuscule, sauf pour "de" et "la")

lambda_function3 = lambda x: re.sub(r"\b\w+\b", lambda x: x.group(0) if x.group(0).lower() in {"de", "la"} else x.group(0).title(), x['Nom_Complet_corr']).strip()
df['Nom_Complet_corr'] = df.apply(lambda_function3, axis=1)
df
```
